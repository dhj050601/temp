## 4. 文件管理
- vi编辑器
    1. 使用j、k、h、l移动光标，可以使用箭头键盘控制替代
    2. 使用“$”光标定位行尾
    3. 使用“^”光标定位行首
    4. 使用“ZZ”实现文本文件的存盘
    
- 文件类型
    可执行文件: ".sh", ".jar"...
- 解压缩tar.gz文件，安装JDK，命令如下：

```python
    # tar zxvf jdk-8u231-linux-x64.tar.gz
    #  mv jdk1.8.0_231 /usr/
    # chown -R hadoop:hadoop jdk1.8.0_231
    #
```

- 配置环境变量，使JDK生效，命令如下：


```python
    $ vi .bash_profile
    $
```

- .bash_profile文件内容如下：

```python
    JAVA_HOME=/usr/jdk1.8.0_231
    PATH=$JAVA_HOME/bin:$PATH

    export JAVA_HOME
    export PATH
```

- 使配置文件生效，命令如下：


```python
    $ source .bash_profile
    $
```
